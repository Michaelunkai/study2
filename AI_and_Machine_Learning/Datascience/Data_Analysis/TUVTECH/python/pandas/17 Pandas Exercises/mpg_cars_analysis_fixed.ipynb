# MPG Cars

### Step 1: Import the necessary libraries

```
# Install necessary dependencies
# !pip install pandas numpy

import pandas as pd
import numpy as np
import requests
```

### Step 2: Import cars1 and cars2 datasets

```
url_cars1 = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
url_cars2 = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

# Download the datasets
cars1_response = requests.get(url_cars1)
cars2_response = requests.get(url_cars2)

with open("cars1.csv", "wb") as file:
    file.write(cars1_response.content)

with open("cars2.csv", "wb") as file:
    file.write(cars2_response.content)

print("Datasets downloaded: 'cars1.csv' and 'cars2.csv'")
```

### Step 3: Assign each dataset to a variable called cars1 and cars2

```
cars1 = pd.read_csv("cars1.csv", header=None)
cars2 = pd.read_csv("cars2.csv", header=None)
```

### Step 4: Fix unnamed blank columns in cars1

```
cars1 = cars1.loc[:, cars1.columns.notnull()]
```

### Step 5: What is the number of observations in each dataset?

```
print(f"Number of observations in cars1: {len(cars1)}")
print(f"Number of observations in cars2: {len(cars2)}")
```

### Step 6: Join cars1 and cars2 into a single DataFrame called cars

```
cars = pd.concat([cars1, cars2], ignore_index=True)
```

### Step 7: Create a random number Series for owners column

```
owners = pd.Series(np.random.randint(15000, 73000, size=len(cars)))
```

### Step 8: Add the column owners to cars

```
cars['owners'] = owners
print(cars.head())
```
