### Step 2: Import the dataset from the given address

```
url = "https://raw.githubusercontent.com/datasets/popular-baby-names/master/data/us-baby-names.csv"
baby_names = pd.read_csv(url)
```

### Step 3: Assign it to a variable called baby_names

```
baby_names.head(10)
```

### Step 4: See the first 10 entries

```
baby_names = baby_names.drop(columns=['Unnamed: 0', 'Id'], errors='ignore')
```

### Step 5: Delete the columns 'Unnamed: 0' and 'Id'

```
gender_counts = baby_names['Gender'].value_counts()
more_male = gender_counts['M'] > gender_counts['F']
print(f"More male names: {more_male}")
```

### Step 6: Is there more male or female names in the dataset?

```
names = baby_names.groupby('Name').agg({'Count': 'sum'})
```

### Step 7: Group the dataset by name and assign to names

```
num_unique_names = names.shape[0]
print(f"Number of different names: {num_unique_names}")
```

### Step 8: How many different names exist in the dataset?

```
most_occurrences = names['Count'].idxmax()
print(f"Name with most occurrences: {most_occurrences}")
```

### Step 9: What is the name with most occurrences?

```
least_occurrences = names[names['Count'] == names['Count'].min()].shape[0]
print(f"Number of names with the least occurrences: {least_occurrences}")
```

### Step 10: How many different names have the least occurrences?

```
median_occurrence = names['Count'].median()
print(f"Median name occurrence: {median_occurrence}")
```

### Step 11: What is the median name occurrence?

```
std_deviation = names['Count'].std()
print(f"Standard deviation of names: {std_deviation}")
```

### Step 12: What is the standard deviation of names?

```
summary = names['Count'].describe()
print(summary)
```
