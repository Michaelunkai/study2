# Scores

### Step 1: Import the necessary libraries

```
# Install necessary dependencies
# !pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2: Create the DataFrame

```
data = {
    'Name': ['Tom', 'Nick', 'Kate', 'John', 'Anna'],
    'Age': [25, 30, 22, 35, 28],
    'Sex': ['Male', 'Male', 'Female', 'Male', 'Female'],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}
scores = pd.DataFrame(data)
print(scores)
```

### Step 3: Create a scatterplot of `preTestScore` and `postTestScore`, with the size determined by age

```
plt.figure(figsize=(10, 6))
plt.scatter(scores['preTestScore'], scores['postTestScore'], s=scores['Age'] * 10, alpha=0.5)
plt.title('Scatterplot of PreTestScore and PostTestScore')
plt.xlabel('PreTestScore')
plt.ylabel('PostTestScore')
plt.show()
```

### Step 4: Create a scatterplot of `preTestScore` and `postTestScore`, with size based on `postTestScore` and color by sex

```
colors = scores['Sex'].map({'Male': 'blue', 'Female': 'red'})
plt.figure(figsize=(10, 6))
plt.scatter(scores['preTestScore'], scores['postTestScore'], s=scores['postTestScore'] * 4.5, c=colors, alpha=0.5)
plt.title('Scatterplot of PreTestScore and PostTestScore (Colored by Sex)')
plt.xlabel('PreTestScore')
plt.ylabel('PostTestScore')
plt.show()
```

### BONUS: Create your own question and answer it

```
# BONUS: What is the correlation between preTestScore and postTestScore?
correlation = scores['preTestScore'].corr(scores['postTestScore'])
print(f'Correlation between PreTestScore and PostTestScore: {correlation:.2f}')
```
